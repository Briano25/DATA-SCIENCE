```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Person': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Num_Children': [2, 3, 1, 4, 0],      # Discrete variable
    'Height_cm': [165.4, 172.8, 158.0, 180.2, 169.5]  # Continuous variable
}

df = pd.DataFrame(data)

# Display the dataset
print(df)

plt.figure(figsize=(6, 4))
plt.bar(df['Person'], df['Num_Children'], color='skyblue')
plt.title('Number of Children per Person (Discrete Variable)')
plt.xlabel('Person')
plt.ylabel('Number of Children')
plt.grid(True, axis='y')
plt.show()
```
