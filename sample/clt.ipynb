```python
import numpy as np
import matplotlib.pyplot as plt

# Create a non-normal (skewed) population
population = np.random.exponential(scale=2.0, size=10000)

# Collect sample means
sample_means = []
for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black')
plt.title('Sampling Distribution of Sample Means (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()
```
