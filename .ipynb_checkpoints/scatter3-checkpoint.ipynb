```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data
df = pd.DataFrame({
    'Age': [25, 30, 35, 40, 45],
    'Salary': [40000, 45000, 50000, 55000, 60000]
})

# Create scatter plot
plt.scatter(df['Age'], df['Salary'])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn scatter plot
sns.scatterplot(x='Age', y='Salary', data=df)

plt.title("Age vs Salary")
plt.show()
```
