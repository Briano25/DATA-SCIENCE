```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
df = pd.DataFrame({
    'Department': ['HR', 'HR', 'IT', 'IT', 'Sales', 'Sales'],
    'Salary': [45000, 47000, 60000, 62000, 52000, 53000]
})

# Draw a box plot of Salary
df.boxplot(column='Salary')

plt.title("Box Plot of Salary")
plt.ylabel("Salary")
plt.show()
```
