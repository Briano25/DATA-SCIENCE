```python
from sklearn.preprocessing import PowerTransformer
import pandas as pd

# Example dataset with negatives and zeros
df = pd.DataFrame({'values': [-5, -2, -1, 0, 1, 2, 3, 5, 10, 20, 50]})

# Apply Yeo-Johnson transform
pt = PowerTransformer(method='yeo-johnson')
df['values_yeojohnson'] = pt.fit_transform(df[['values']])

# Calculate skewness
skew_before = df['values'].skew()
skew_after = df['values_yeojohnson'].skew()

print("Skewness before Yeo-Johnson:", skew_before)
print("Skewness after Yeo-Johnson:", skew_after)
print(df.head(10))
```

```python
import matplotlib.pyplot as plt 
plt.subplot(1,2,1)
plt.hist(df['values'], bins=10, edgecolor='black')
plt.title("Original Data (Skewed)")
plt.xlabel("Values")
plt.ylabel("Frequency")

plt.subplot(1,2,2)
plt.hist(df['values_yeojohnson'], bins=10, edgecolor='black')
plt.title("Yeo-Johnson Transformed Data (Reduced Skew)")
plt.xlabel("Transformed Values")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
