```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset("titanic")

# Convert 'sex' to numeric
df['sex_encoded'] = df['sex'].map({'male': 0, 'female': 1})

# Create correlation matrix
corr = df[['survived', 'pclass', 'age', 'fare', 'sex_encoded']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
ax = sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Add arrows (text) to highlight comparisons
#  shows relationship between two variables

# survived vs sex_encoded
ax.annotate('', xy=(0.5, 4.2), fontsize=16, color='black')
ax.text(0.5, 4.4, "Survived vs Gender", fontsize=9, ha='center')

# fare vs pclass
ax.annotate('', xy=(3.5, 1.2), fontsize=16, color='black')
ax.text(3.5, 1.4, "Fare vs Class", fontsize=9, ha='center')

plt.title("Annotated Correlation Heatmap - Titanic")
plt.tight_layout()
plt.show()
```
