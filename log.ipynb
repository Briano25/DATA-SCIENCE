```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample skewed data
data = pd.Series([1, 2, 3, 4, 5, 10, 20, 50, 100, 200, 500])

# Before transform
plt.hist(data, bins=10)
plt.title("Original Data")
plt.show()

# Apply natural log
data_log = np.log(data)

# After transform
plt.hist(data_log, bins=10)
plt.title("Log-Transformed Data")
plt.show()
```
